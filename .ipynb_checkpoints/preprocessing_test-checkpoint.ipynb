```python
import pandas as pd
import numpy as np
import json
import timeit
```

```python
path = "data/kkosek.json"
f = open(path)
#df = pd.read_json(path)
obj = json.load(f)
```

```python
start_time = timeit.default_timer()
len_workout = len(obj["lift_records"][0])
len_workouts = len(obj["lift_records"])
columns = ["exerciseName","setCount","avgReps","avgWeights","RPE"]
df = pd.DataFrame(columns = columns)
for i in range(0,len_workouts):
    len_workout = len(obj["lift_records"][i])
    for j in range(0,len_workout):
        set_count = len(obj["lift_records"][i][j]["weights"])
        name = obj["lift_records"][i][j]["exerciseName"]
        avg_reps = np.mean(obj["lift_records"][i][j]["reps"])
        avg_weights = np.mean(obj["lift_records"][i][j]["weights"])
        rpe = np.mean(obj["lift_records"][i][j]["rpe"])
        date = obj["lift_records"][i][j]["date"]
        df = pd.concat([df,pd.DataFrame({
            "exerciseName":name,
            "setCount":set_count,
            "avgReps":avg_reps,
            "avgWeights":avg_weights,
            "RPE":rpe,
        }, index=[date])])
display(df)
end_time = timeit.default_timer()
print("Execution time:", end_time - start_time)
```
